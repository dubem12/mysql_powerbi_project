```python

import pandas as pd
```

```python
df = pd.read_csv('grouped_melted.csv')

df
```

convert all rows into a list of tuples

```python
y= []

for i in range (len(df)):
    x= tuple(df.iloc[i])
    y.append(x)

y
```

```python
#save the list of tuples to a file

file = open('product.txt', 'w')

for tuple in y:
    file.write(str(tuple) + ',' + '\n')

file.close()
```
